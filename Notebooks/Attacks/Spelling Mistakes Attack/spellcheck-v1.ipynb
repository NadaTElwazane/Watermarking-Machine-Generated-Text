```python
import pandas as pd
from tqdm import tqdm
from transformers import pipeline

# Assuming the CSV with "Modified Text" already exists

# Load the data from the CSV file
data = pd.read_csv("/kaggle/input/spelling-mistakes-dataset/mispelled_text.csv")

# Load the text-to-text generation pipeline for spelling correction
fix_spelling = pipeline("text2text-generation", model="oliverguhr/spelling-correction-english-base")

# Create an empty list to store corrected text
corrected_text = []

# Loop through the "Modified Text" column and apply spelling correction
for text in tqdm(data["Modified Text"]):
    correction = fix_spelling(text, max_length=2048)[0]["generated_text"].strip()
#     print(correction)
    corrected_text.append(correction)

# Add the corrected text as a new column in the dataframe
data["Corrected Text"] = corrected_text

# Save the updated dataframe with corrected text
data.to_csv("corrected_text.csv", index=False)

print("Spelling correction complete! Corrected text added to 'corrected_text.csv'")
```
