```python
# Read data from CSV using pandas
data = pd.read_csv("/kaggle/input/watermarked-and-unwatermarked-text/data.csv")
# Filter data (assuming "label" is the column for watermarked/unwatermarked)
watermarked_data = data[data["label"] == "watermarked"].sample(1000, random_state=42)  # Seed for reproducibility
unwatermarked_data = data[data["label"] == "unwatermarked"].sample(1000, random_state=42)

# Combine filtered data
combined_data = pd.concat([watermarked_data, unwatermarked_data], ignore_index=True)
combined_data = combined_data.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle and reset index


combined_data.to_csv("data.csv", index=False)

# Manual Split (approximately equal parts, last part might be slightly bigger)
split_point1 = int(len(combined_data) / 3)
split_point2 = split_point1 * 2

part1 = combined_data.iloc[:split_point1]  # First part
part2 = combined_data.iloc[split_point1:split_point2]  # Second part
part3 = combined_data.iloc[split_point2:]  # Third part (might be slightly bigger)

number_of_rows = len(part1)

# Print the results
print(f"The CSV file has {number_of_rows} rows.")
```
