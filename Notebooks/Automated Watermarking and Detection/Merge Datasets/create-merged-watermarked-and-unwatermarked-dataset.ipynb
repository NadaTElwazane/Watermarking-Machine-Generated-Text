```python
import pandas as pd

# Define file paths (replace with your actual paths)
watermarked_file = "/kaggle/input/watermarked-text/Watermarked Dataset.csv"
unwatermarked_file = "/kaggle/input/unwatermarked-text/unwatermarked_texts.csv"

# Read CSV files into DataFrames
watermarked_df = pd.read_csv(watermarked_file)
unwatermarked_df = pd.read_csv(unwatermarked_file,nrows=2500)

# Create label columns
watermarked_df['label'] = 'watermarked'
unwatermarked_df['label'] = 'unwatermarked'

# Merge DataFrames (assuming compatible column structures)
merged_df = pd.concat([watermarked_df, unwatermarked_df])

# Save the merged DataFrame to a new CSV file (optional)
merged_df.to_csv("data.csv", index=False)

print("CSV files merged with label column!")
```
