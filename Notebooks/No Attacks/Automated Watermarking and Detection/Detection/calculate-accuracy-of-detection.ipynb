```python
import pandas as pd

def calculate_accuracy(file_path):
    """
    Calculates accuracy from a CSV file containing 'label' and 'actual_label' columns.

    Args:
      file_path (str): Path to the CSV file.

    Returns:
      float: Accuracy (between 0 and 1).
    """

    # Read the CSV file into a Pandas DataFrame
    df = pd.read_csv(file_path)

    # Calculate the number of correct predictions
    correct_predictions = (df['label'] == df['actual_label']).sum()

    # Calculate the total number of predictions
    total_predictions = len(df)

    # Calculate accuracy
    accuracy = correct_predictions / total_predictions

    return accuracy


file_path = "/kaggle/input/p-value-result-no-attacks/p-values no attack.csv"  # Replace with your actual file path
accuracy = calculate_accuracy(file_path)

print(f"Accuracy: {accuracy*100:.3f}")
```
